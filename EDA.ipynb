```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('Equifax_pdl_US_2023.csv')
df.drop_duplicates()
print(df.head(5))
print(df.shape)
print(list(df))
```

```python
print(df.describe(include='all'))
print('Check NaN:')
df.isnull().sum()
```

The 'client_id' and 'application_date' fields were ignored due to their uninformative nature

```python
df.drop(['client_id', 'application_date'], axis=1 , inplace=True)
```

```python
plt.boxplot(df['age'])
plt.title('age')
plt.show()
```

```python
frequency = df['default_status'].value_counts()
values = list(set(df['default_status'].to_list()))
values = list(map(lambda x: str(x), values))
plt.bar(values, frequency)
plt.title('default_status')
plt.show()
```

```python
frequency = df['education_level'].value_counts()
values = list(set(df['education_level'].to_list()))
plt.bar(values, frequency)
plt.title('education_level')
plt.legend(['ACD', 'GRD', 'PGR', 'SCH', 'UGR'],
           ["Associate's Degree", 'Graduate Degree',
            'Postgraduate Degree', 'School Level Education', 'Undergraduate Degree'])
plt.show()
```
